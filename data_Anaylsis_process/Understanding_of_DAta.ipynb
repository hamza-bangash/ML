# 2.EDA

### Before Starting EDA we should understand Data Type

        There are two Types
         1.Numerical    : Numbers for example age,height etc.
         2.Categorical  : for example Gender,ratings etc
             i.Nominal: order does not matter for example colors red , blue, yellow so it does not mean yellow is good than red vice verce
             ii.ordinal: order does matter for example rating best , average , worst  so best > average > worst

### Univaraite 

####      1. Categorical Data

    a.CountPlot

```python
sns.countplot(df['Age'])
```

```python
## different method
df['Age'].value_counts().plot(kind='bar')
## count plot actually do the same work first it calculate the counts then crate  bar plot
```

    b.PieChart

```python
df['Gender'].value_counts().plot(kind='pie',autopct='%0.2f%%')
```

####      2.Numerical Data

    a.Histogram

```python
plt.hist(df['Product_Category_1'],bins=15)
```

    b.DistPlot

    c.BoxPlot

```python
sns.boxplot(df['Purchase'])
```

### Bivariate and Multivariate
    hue , style  parameters give us multivariate

#### 1.Numerical - Numerical cols

    a.Scatterplot and linePlot

```python
sns.scatterplot(x=df['Product_Category_1'], y=df['Purchase'],hue=df["Gender"]) # Using hue parameter give us multivariate 
```

#### 2.Numerical - Categorical cols

    a.Bar Plot

```python
sns.barplot(x=df['Gender'],y=df['Purchase'],hue=df['City_Category']) # You can also use hue parameter to achive multivarite analysis
```

    b.Displot

```python
sns.displot(x=df['Product_Category_3'])
sns.displot(x=df['Product_Category_1']) # Read about when applying it the data does write during theis time that wy its not worinking
```

    c.BoxPlot

```python
sns.boxplot(x=df['City_Category'],y=df['Product_Category_1'],hue=df['Gender'])
```

#### 3.Categorical - Categorical cols

    a.HeatMap

```python
sns.heatmap(pd.crosstab(df['Age'],df['Gender']))
```

    b.Cluster

```python
sns.clustermap(pd.crosstab(df['Age'],df['Gender']))
```

    c.PairPlot

```python
sns.pairplot(df)
```
