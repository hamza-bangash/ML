## Cleaning and Preparing the Data for model trainging

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#importing the dataset
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
## import the test data
df_test = pd.read_csv('test.csv')
df_test.head()
```

```python
#merge train and test data
df = pd.concat([df_train,df_test],axis=0,ignore_index=True)
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#Handle categorical Gender
df.Gender = df.Gender.map({'F':0,'M':1})
```

```python
df.head()
```

```python
## Handle Categorical Age
df.Age.unique()
```

```python
## pd.get_dummies(df['Age'],drop_first=True)
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
```

```python
df.head()
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city
```

```python
pd.concat([df,df_city],axis=1)
df.head()
```

```python
df.drop(['City_Category'],axis=1,inplace=True)
df.head()
```

```python
df.isnull().sum()
```

```python
##focus on replacing missing values
df.Product_Category_2.unique()
```

```python
df.Product_Category_2.value_counts()
```

```python
#Replace the missing value with mode
df.Product_Category_2.mode() #so this will give us mode of product which is 8.0 in the form series. bacz it return series there multiple mode value (same most repeated value).
df.Product_Category_2.mode()[0] # by this we will get first value
```

```python
df.Product_Category_2=df.Product_Category_2.fillna(df.Product_Category_2).mode()[0]
```

```python
df.Product_Category_3=df.Product_Category_3.fillna(df.Product_Category_3).mode()[0]
```

```python
df.isnull().sum()
```

```python
# Stay_in_Current_City_Years

df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df.info()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
df[['B','C']]=df[['B','C']].astype(int)
```
