## 2.Cleaning Data

    1.droping columns 
        following are the columns
        a.county bcuz 100% missing values
        b.region url bucz region columns present so don't need to region url
        

```python
drop_columns=['county','region_url']
df = df.drop(columns=drop_columns)
```

    2.Handling Missing values

```python
##Handling size columns
df['size'].unique()
df['size']=df['size'].fillna(df['size'].mode()[0])
```

```python
## Handling Cylinder columns
##first it should remove from mix datatypes to int
df['cylinders'] = df['cylinders'].str.extract(r'(\d+)')

## Second handling missing values
df['cylinders'] = df['cylinders'].fillna(df['cylinders'].mode()[0])
```

```python
## Handling missing value of  all categorical cols with mode
categorical_cols = df.select_dtypes(include=['object']).columns
categorical_cols

for col in categorical_cols:
    df[col]=df[col].fillna(df[col].mode()[0])
```

```python
numerical_cols = df.select_dtypes(include=['number']).columns

for col in numerical_cols:
    df[col]=df[col].fillna(df[col].mean())
```

```python
df['posting_date'] = pd.to_datetime(df['posting_date'],utc=True)
```
