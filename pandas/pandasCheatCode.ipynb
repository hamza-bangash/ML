```python
teams = vk.groupby('bowling_team')
teams.head()
```

```python
teams['batsman_runs'].sum().sort_values(ascending=False).head()
```

```python
def run_scored(batsman_name):
    mask = data['batsman']==batsman_name
    vk = data[mask]
    teams = vk.groupby('bowling_team')
    return (teams['batsman_runs'].sum().sort_values(ascending=False).head(1))
    
```

```python
run_scored('MS Dhoni')
```

isin() function

```python
mask = data['over']>15
new_data = data[mask]
```

```python
batsman= new_data.groupby('batsman')['batsman_runs'].count()
mask = batsman>=200
batsman_list=batsman[mask].index.tolist()
```

```python
mask = data['batsman'].isin(batsman_list)
```

```python
final = data[mask]
```

```python
runs = final.groupby('batsman')['batsman_runs'].sum()
balls = final.groupby('batsman')['batsman_runs'].count()
```

```python
sr = (runs/balls)*100
```

```python
sr
```

merge() function 

```python
match = pd.read_csv('matches.csv')
```

```python
delivery = pd.read_csv('deliveries.csv')
```

```python
new=delivery.merge(match,left_on='match_id',right_on='id')
```

```python
new.groupby(['season','batsman'])['batsman_runs'].sum().sort_values(ascending=False).reset_index().drop_duplicates(subset='season',keep='first').sort_values('season',ascending=False)
```

Pivot Table

```python
mask = delivery['batsman_runs']==6
six = delivery[mask]
```

```python
pt=six.pivot_table(index='over',columns='batting_team',values='batsman_runs',aggfunc='count')
```

```python
import seaborn as sns
```

```python
sns.heatmap(pt)
```
