```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("train.csv")
data = data['Age']
```

```python
# Extract the 'Age' column
age_data = data['Age']

# Calculate mean, median, and mode
mean_age = age_data.mean()
median_age = age_data.median()
mode_age = age_data.mode()[0]

# Create a histogram
plt.figure(figsize=(10, 6))
sns.histplot(age_data, bins=10, kde=False, color='skyblue', stat='density')

# Add lines for mean, median, and mode
plt.axvline(mean_age, color='red', linestyle='dashed', linewidth=2, label='Mean: {:.2f}'.format(mean_age))
plt.axvline(median_age, color='green', linestyle='dashed', linewidth=2, label='Median: {:.2f}'.format(median_age))
plt.axvline(mode_age, color='blue', linestyle='dashed', linewidth=2, label='Mode: {:.2f}'.format(mode_age))

# Add titles and labels
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```

```python
# Convert to DataFrame
df = pd.DataFrame(data)

# Create a box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Age'], color='lightblue')

# Add titles and labels
plt.title('Box Plot of Passenger Ages')
plt.xlabel('Age')
plt.grid()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# Sample Data
mean = 30
std_dev = 5
variance = std_dev ** 2

# Create a range of x values
x = np.linspace(10, 50, 100)
# Calculate the normal distribution for each x value
y = norm.pdf(x, mean, std_dev)

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(x, y, color='blue', label='Normal Distribution\nMean: {}\nVariance: {}\nStd Dev: {}'.format(mean, variance, std_dev))

# Highlight the mean
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label='Mean: {}'.format(mean))

# Highlight standard deviations
plt.axvline(mean - std_dev, color='orange', linestyle='dashed', linewidth=2, label='Mean - 1 Std Dev')
plt.axvline(mean + std_dev, color='orange', linestyle='dashed', linewidth=2, label='Mean + 1 Std Dev')
plt.axvline(mean - 2*std_dev, color='green', linestyle='dashed', linewidth=2, label='Mean - 2 Std Dev')
plt.axvline(mean + 2*std_dev, color='green', linestyle='dashed', linewidth=2, label='Mean + 2 Std Dev')

# Fill the areas under the curve for the standard deviations
plt.fill_between(x, y, 0, where=(x > mean - std_dev) & (x < mean + std_dev), color='lightblue', alpha=0.5, label='Area within 1 Std Dev (~68%)')
plt.fill_between(x, y, 0, where=(x > mean - 2*std_dev) & (x < mean + 2*std_dev), color='lightgreen', alpha=0.3, label='Area within 2 Std Devs (~95%)')

# Add titles and labels
plt.title('Normal Distribution Curve with Standard Deviations and Variance')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid()

# Show the plot
plt.show()
```
