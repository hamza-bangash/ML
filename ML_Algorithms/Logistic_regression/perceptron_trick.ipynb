```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import make_classification 
x,y = make_classification(n_samples=100,n_features=2,n_informative=1,
                          n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
        
```

```python
def perceptron(x,y):
    x=np.insert(x,0,1,axis=1) # this will include in x matric x0 it will add first column of value 1
    weights = np.ones(x.shape[1]) # it will create array of one of length equal number column
    lr=0.5

    for i in range(1000):
        j = np.random.randint(0,100) # generate random value
        y_hat = step(np.dot(x[j],weights)) # this will multiple random row with coffients [x0,x1,x2].transpose([w0,w1,w2])
        weights = weights + lr*(y[j] - y_hat)* x[j]

    return weights[0],weights[1:]
                     
    
```

```python
def step(z):
    value = 0
    if(z>0):
        value =1
    return value
```

```python
intercept_,coef_ = perceptron(x,y)
```

```python
print(coef_)
print(intercept_)
```

```python
b = -(intercept_/coef_[1])
m= -(coef_[0]/coef_[1])   # x and y cofficient is coef_[0] and coef_[1] which is equal to slope m = -a/b
```

```python
x_input = np.linspace(-1,0,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```
