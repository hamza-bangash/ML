- ##### Implementaion of Standardization

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.sample(5)
```

```python
from sklearn.model_selection  import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df[['Age','EstimatedSalary']],df['Purchased'],test_size=0.3,random_state=0)

print(x_train.shape)
print(y_train.shape)
```

```python
#StandardScaler

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x_train)  # The fit function is used to learn to calculate mean and standard varaince Noted: we only sent x_train 
x_train_scaled = scaler.transform(x_train) # in transform we transform both x_trian and x_test
x_test_scaled = scaler.transform(x_test)
```

```python
# Now after scale

x_train_scaled = pd.DataFrame(x_train_scaled,columns=x_train.columns) # chaning from array to data frame
x_test_scaled = pd.DataFrame(x_test_scaled,columns=x_test.columns)
```

```python
# before scale
np.round(x_train.describe(),1)
```

```python
# after scale

np.round(x_train_scaled.describe(),1)
```

```python
fig,(ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.scatter(x_train['Age'],x_train['EstimatedSalary'])
ax1.set_title("Before Scaling")

ax2.scatter(x_train_scaled['Age'],x_train_scaled['EstimatedSalary'],color='red')
ax1.set_title("After Scaling")

plt.show()
            
```

```python
fig , (ax1,ax2) = plt.subplots(ncols=2,figsize=(12,5))

ax1.set_title('Before Scaling')
sns.kdeplot(x_train['Age'],ax=ax1)
sns.kdeplot(x_train['EstimatedSalary'],ax=ax1)

ax2.set_title("After Scaling")
sns.kdeplot(x_train_scaled['Age'],ax=ax2)
sns.kdeplot(x_train_scaled['EstimatedSalary'],ax=ax2)

plt.show()
              
```

```python
# using Logistic regression to check the perfomace using scaling
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression() # creating object of logistic alogorith
lr_scaled = LogisticRegression() # creation second object

lr.fit(x_train,y_train)
lr_scaled.fit(x_train_scaled,y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred_scaled = lr_scaled.predict(x_test_scaled)
```

```python
from sklearn.metrics import accuracy_score

print("Acctual",accuracy_score(y_test,y_pred))
print("Scaled",accuracy_score(y_test,y_pred_scaled))
      
```

2. #### Normalization

        -All values are squeezed between 0 and 1 (or another range like -1 to 1).
        -Preserves relationships between values but compresses outliers.
        -Best Used When:
        -Data has different ranges or units (e.g., heights in cm vs. weights in kg).
        -You need bounded values (e.g., neural networks work better with normalized inputs).

    1.There are different type of normalization
        1.minmax scaling
        2.mean normalization
        3.max normalizatiion
        4.robust normalization

MinMax Scaling formula:
        xi = (xi-xmin)/(xmax -xmin)

```python
df = pd.read_csv("wine_data.csv",header=None,usecols=[0,1,2])
```

```python
color_dict = {1:'red',3:'green',2:'blue'}
sns.scatterplot(df['Alcohol'],df['Malic acid'],hue=df['Class label'],palette=color_dict)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Class label',axis=1),df['Class label'],test_size=0.3,random_state=0)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()

scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```
