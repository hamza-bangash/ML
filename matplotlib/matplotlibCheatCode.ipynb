Working With Data csv File

```python
import csv
from collections import Counter
```

```python
with open('data.csv') as csv_file: # it will open the file after that it will colse atuomatically bcz using with
    csv_reader = csv.DictReader(csv_file)  #convert the csv file into dictionary
           
    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

languages= []
popularity = []

for item in language_counter.most_common():
    languages.append(item[0])
    popularity.append(item[1])

print(languages)
print(popularity)
```

```python
plt.barh(languages,popularity)

plt.title('Most popular languages')
plt.xlabel('Programming Languages')
plt.ylabel('Number of People Who use')
```

Stack Plot

```python
minutes = [1,2,3,4,5,6,7,8,9]

player1 = [2,4,5,7,1,3,2,2,1]
player2 = [3,2,5,6,5,9,6,3,2]
player3 = [6,1,2,1,1,7,8,3,2]

players = ['player1','player2','player3']

plt.stackplot(minutes,player1,player2,player3,labels=players)
plt.title('Stack plot')
plt.tight_layout()
plt.legend()
```

Line plot

```python
import pandas as pd
data = pd.read_csv('languages.csv')
```

```python
data.head()
```

```python
ages = data['Age']

allDevs = data['All_Devs']

pyDevs = data['Python']

jsDevs = data['JavaScript']
```

```python
plt.plot(ages,allDevs,label='All Devs')
plt.plot(ages,pyDevs,label='Py Devs')
plt.plot(ages,jsDevs,label='Js Devs')

plt.tight_layout()
plt.xlabel('Ages')
plt.ylabel('Salary')
plt.title('Salary of languages User')


medain_salary = 57287
plt.fill_between(ages, pyDevs,medain_salary,where=(pyDevs>=medain_salary),label='Above Average',
                interpolate=True, alpha=0.25)   #This method will fill the entire below the given line
plt.fill_between(ages, pyDevs,medain_salary,where=(pyDevs<medain_salary),label='below Average',
                interpolate=True ,alpha=0.25)   #This method will fill the entire below the given line

plt.legend()
```
