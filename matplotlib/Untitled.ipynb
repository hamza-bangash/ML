```python
import numpy as np
import matplotlib.pyplot as plt

# Raw data
people = ['A', 'B', 'C', 'D', 'E']
height = [160, 170, 175, 180, 190]
weight = [55, 65, 75, 85, 95]

# Calculate mean and standard deviation
mean_height = np.mean(height)
std_height = np.std(height)
mean_weight = np.mean(weight)
std_weight = np.std(weight)

# Standardize the data
z_height = [(h - mean_height)/std_height for h in height]
z_weight = [(w - mean_weight)/std_weight for w in weight]

# Create plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1: Raw Data (Different Scales)
ax1.plot(people, height, 'o-', color='blue', label='Height (cm)')
ax1.set_ylabel('Height (cm)', color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_ylim(150, 200)

ax1b = ax1.twinx()
ax1b.plot(people, weight, 'o-', color='red', label='Weight (kg)')
ax1b.set_ylabel('Weight (kg)', color='red')
ax1b.tick_params(axis='y', labelcolor='red')
ax1b.set_ylim(50, 100)

ax1.set_title('Raw Data (Different Scales)')
ax1.grid(True)

# Plot 2: Standardized Data (Same Scale)
ax2.plot(people, z_height, 'o-', color='blue', label='Z-Height')
ax2.plot(people, z_weight, 'o-', color='red', label='Z-Weight')
ax2.set_ylabel('Z-Scores')
ax2.set_title('Standardized Data (Same Scale)')
ax2.legend()
ax2.grid(True)

plt.tight_layout()
plt.show()
```
