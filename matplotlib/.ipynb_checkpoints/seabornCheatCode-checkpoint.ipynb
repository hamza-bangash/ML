# Starting 

**Importing libraries**

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

**Loading Data from Seeborn**

```python
tips = sns.load_dataset('tips')
tips.head()
```

## 1)Relation Plot 
it show the relation between the entities how they are relate to eachother 

- Scatter Plot

```python
sns.scatterplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size')   #using axes function to develop scatter plot
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',hue='sex',style='time',size='size',kind='scatter') #Using Figure function to develop scatter plot
```

- line Plot

```python
#the difference line plot and scatter is that line plot show the contuntity by using line to join that points and line better to use if there is a time like year,days etc
```

```python
gap = px.data.gapminder() # fetching data
gap.tail()
```

```python
pakistan = gap[gap['country']=='Pakistan']
pakistan
```

```python
sns.relplot(kind='line',data=pakistan,x='year',y='lifeExp')
```

```python
temp_df=gap[gap['country'].isin(['Brazil','Pakistan','Germany'])]
temp_df
```

```python
sns.relplot(kind='line',data=temp_df,x='year',y='lifeExp',hue='country',style='continent',size='continent')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex') #col attributes used to divide the graph based categorical dat
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='smoker') #row attributes used to divide the graph based categorical dat
```

## 2)Distribution Plot
it show how much over data are divided it show the show the distribution

it is use for
1 univaraite analysis
2 used for find out the distribution
3 range of the observation
4 cental tendency
5 is the data bimodal
6 are there outliers

- figure level -> displot
- axis level  ->  histplot,kdeplot,rugplot

- **Histgram plot**
   it show the count of data using bin

```python
sns.displot(data=tips,x='total_bill',kind='hist')
```

```python
sns.displot(data=tips,x='total_bill',kind='hist',bins=20) #bins  attributs used to increase or decrease number of bins
```

```python
sns.displot(data=tips,x='day',kind='hist') # actually for this count plot is used
```

```python
sns.displot(data=tips,x='tip',kind='hist',hue='sex',element='step')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
sns.displot(data=titanic,x='age',kind='hist',element='step',col='survived')
```

- **kde plot**
  its works like hist plot but the only difference its show the contuntity
  - Actually it show the probablity

```python
sns.displot(data=tips,x='total_bill',kind='kde',hue='sex') 
```

- **Rug Plot** 
  - plot marginal distributions by drawing ticks along the x and y axes.
  - this plot is intended to complement other plots by showing the location of individual obsevations in an unobtrusive way.

```python
#using axis leve function
sns.kdeplot(data=tips,x='total_bill')
sns.rugplot(data=tips,x='total_bill')
```

## 3)Matrix 

- Heatmap
  plot rectangle data as a color-encode matrix

```python
# making data clear 
temp_df=gap.pivot(index='country',columns='year',values='lifeExp')
temp_df
```

```python
plt.figure(figsize=(15,10)) # This function used to increase the size of graph
sns.heatmap(temp_df) #axis level function for heatmap figure function does not exist
```

```python
#annot attribute will show the number on each
#linewidth will increase width in each protation

temp_df=gap[gap['continent']=='Asia'].pivot(index='country',columns='year',values='lifeExp')
plt.figure(figsize=(15,10))
sns.heatmap(temp_df,annot=True,linewidth=0.5)
```

- **Clustermap**

```python
iris = px.data.iris()
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```

## 4) Categorical Plot

**There are three categories in categorical plot**
 - categorical scatter plot
    - stripplot
    - swarmplot
 - categorical distribution plots
    - boxplot
    - violinplot
 - categorical Estimate plot -> for central tendency
    - barplot
    - pointplot
    -countplot

Figure level Function -> catplot

**Scatter Plot** 
 - Strip Plot

```python
sns.catplot(data=tips,x='day',y='total_bill')
```

```python
#jitter attributes will show you the real data points
sns.catplot(data=tips,x='day',y='total_bill',jitter=False)
```

- swarmplot

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='swarm')
```

- **categorical distrubation plot**

- boxplot

```python
sns.catplot(data=tips,x='total_bill',kind='box')
```

- violinplot
  - violinplot is the combination of kde and box 

```python
sns.catplot(data=tips,x='total_bill',kind='violin')
```

- **Categorical Estimate Plot**

- barPlot

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='bar')
```

- point plot

```python
sns.catplot(data=tips,x='sex',y='total_bill',kind='point')
```

- count plot

```python
sns.catplot(data=tips,x='sex',kind='count')
```

## 5) Regression Plot

- residual plot
- **Both function work the same but difference  one is axis and one is fugure level function**
- regplot -> axis level
- lmplot  -> figure level

```python
sns.regplot(data=tips,x='total_bill',y='tip')
```

```python
sns.lmplot(data=tips,x='total_bill',y='tip',hue='sex') # hue parameter is avaible in lmplot but not in reg plot
```

- residual plot  -> give us the error of regression 

```python
sns.residplot(data=tips,x='total_bill',y='tip')
```

## 6) Multigrids Plot

- Facetplot
- jointplot
- pairplot

```python
g = sns.FacetGrid(data=tips,col='day',row='time')
g.map(sns.violinplot,'sex','total_bill')
```

- pairplot -> it will give you scatter plot of column of data pair by pair

```python
sns.pairplot(iris,hue='species') 
```

- pairGrid is same is pairPlot but its a figure level

```python
g=sns.PairGrid(iris)
g.map(sns.scatterplot)
```

- JointGrid vs Joinplot

```python
# JoinPlot

sns.jointplot(data=tips,x='total_bill',y='tip',kind='reg')
```

```python
g = sns.JointGrid(data=tips,x='total_bill',y='tip')
g.plot(sns.scatterplot,sns.boxplot)
```
